```python
import os
import json
import pandas as pd
import numpy as np
import mysql.connector as sql
import streamlit as st
import plotly.express as px
```

```python
# Dataframe creation of aggregated transaction 
# aggregated transcations for all states from phonepe clone data 

path1 = "C:/CapstoneProject2/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list = os.listdir(path1)

# getting needed informations from the json file as dictionary

columns1 = {'States': [], 'Year': [], 'Quarters':[], 'Transaction_type': [], 'Transaction_count':[], 'Transaction_amount':[] }

# getting all states from the file

for state in agg_trans_list:
    current_state = path1+state+"/"
    agg_year_list = os.listdir(current_state)

# getting the year for all  the states
    
    for year in agg_year_list:
        current_year = current_state + year +"/"
        agg_file_list = os.listdir(current_year)

 # reading the json file

        for file in agg_file_list:
            current_file = current_year + file
            data = open(current_file, 'r')
            first_data = json.load(data)

            for i in first_data['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columns1['Transaction_type'].append(name)
                columns1['Transaction_count'].append(count)
                columns1['Transaction_amount'].append(amount)
                columns1['States'].append(state)
                columns1['Year'].append(year)
                columns1['Quarters'].append(int(file.strip('.json')))
df_agg_trans = pd.DataFrame(columns1)

    
```

```python
# Dataframe creation of aggregated user 

path2 = "C:/CapstoneProject2/pulse/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path2)

# column creation for the json file

columns2 = {'States':[], 'Year':[], 'Quarters':[], 'Brands':[], 'Count': [], 'Percentage':[] }

# getting all states from the file 

for state in agg_user_list:
    current_state = path2 + state + "/"
    agg_year_list = os.listdir(current_state)

    
# getting the year for all  the states
    
    for year in agg_year_list:
        current_year = current_state + year + "/"
        agg_file_list = os.listdir(current_year)

 # reading the json file
        
        for file in agg_file_list:
            current_file = current_year + file
            data = open(current_file,'r')
            
            second_data = json.load(data)
            try:
                for i in second_data['data']['usersByDevice']:

                    brand_name = i['brand']
                    count = i['count']
                    percentage = i['percentage']
                    columns2['Brands'].append(brand_name)
                    columns2['Count'].append(count)
                    columns2['Percentage'].append(percentage)
                    columns2['States'].append(state)
                    columns2['Year'].append(year)
                    columns2['Quarters'].append(int(file.strip('.json')))
            except:
             pass    
            df_agg_user = pd.DataFrame(columns2)


    
```

```python
# Dataframe creation of map transactions 


path3 = "C:/CapstoneProject2/pulse/data/map/transaction/hover/country/india/state/"
map_trans_list = os.listdir(path3)

# column creation for the json file

columns3 = {'States':[], 'Year':[], 'Quarters':[], 'District':[], 'Transaction_Count': [], 'Transaction_Amount':[] }

# getting all states from the file 

for state in map_trans_list:
    current_state = path3 + state + "/"
    map_year_list = os.listdir(current_state)

    
# getting the year for all  the states
    
    for year in map_year_list:
        current_year = current_state + year + "/"
        map_file_list = os.listdir(current_year)

 # reading the json file
        
        for file in map_file_list:
            current_file = current_year + file
            data = open(current_file,'r')
            
            third_data = json.load(data)

            for i in third_data['data']['hoverDataList']:

                district = i['name']
                count = i['metric'][0]['count']
                amount = i['metric'][0]['amount']
                columns3['District'].append(district)
                columns3['Transaction_Count'].append(count)
                columns3['Transaction_Amount'].append(amount)
                columns3['States'].append(state)
                columns3['Year'].append(year)
                columns3['Quarters'].append(int(file.strip('.json')))

            df_map_trans= pd.DataFrame(columns3)
           
```

```python
# Dataframe creation of map users for all states

path4 = "C:/CapstoneProject2/pulse/data/map/user/hover/country/india/state/"
map_user_list = os.listdir(path4)

# column creation for the json file

columns4 = {'States':[], 'Year':[], 'Quarters':[], 'District':[], 'Registered_Users': [], 'AppOpens':[] }

# getting all states from the file 

for state in map_user_list:
    current_state = path4 + state + "/"
    map_year_list = os.listdir(current_state)

    
# getting the year for all  the states
    
    for year in map_year_list:
        current_year = current_state + year + "/"
        map_file_list = os.listdir(current_year)

 # reading the json file
        
        for file in map_file_list:
            current_file = current_year + file
            data = open(current_file,'r')
            
            fourth_data = json.load(data)
            
            for i in fourth_data['data']['hoverData'].items():

                    district = i[0]
                    registeredUsers = i[1]['registeredUsers']
                    appOpens = i[1]['appOpens']
                    columns4['District'].append(district)
                    columns4['Registered_Users'].append(registeredUsers)
                    columns4['AppOpens'].append(appOpens)
                    columns4['States'].append(state)
                    columns4['Year'].append(year)
                    columns4['Quarters'].append(int(file.strip('.json')))

    df_map_user= pd.DataFrame(columns4)

        
```

```python
# Dataframe creation of top transactions for all states

path5 = "C:/CapstoneProject2/pulse/data/top/transaction/country/india/state/"
top_trans_list = os.listdir(path5)

# column creation for the json file

columns5 = {'States':[], 'Year':[], 'Quarters':[], 'Pincode':[], 'Transaction_Count': [], 'Transaction_Amount':[] }

# getting all states from the file 

for state in top_trans_list:
    current_state = path5 + state + "/"
    top_year_list = os.listdir(current_state)

    
# getting the year for all  the states
    
    for year in top_year_list:
        current_year = current_state + year + "/"
        top_file_list = os.listdir(current_year)

 # reading the json file
        
        for file in top_file_list:
            current_file = current_year + file
            data = open(current_file,'r')
            
            fifth_data = json.load(data)

            for i in fifth_data['data']['pincodes']:
                    
                    name = i['entityName']
                    count = i['metric']['count']
                    amount = i['metric']['amount']
                    columns5['Pincode'].append(name)
                    columns5['Transaction_Count'].append(count)
                    columns5['Transaction_Amount'].append(amount)
                    columns5['States'].append(state)
                    columns5['Year'].append(year)
                    columns5['Quarters'].append(int(file.strip('.json')))

df_top_trans= pd.DataFrame(columns5)
            
                    
```

```python
# Dataframe creation of top user for all states

path6 = "C:/CapstoneProject2/pulse/data/top/user/country/india/state/"
top_user_list = os.listdir(path6)

# column creation for the json file

columns6 = {'States':[], 'Year':[], 'Quarters':[], 'Pincode':[],'Registered_Users':[] }

# getting all states from the file 

for state in top_user_list:
    current_state = path6 + state + "/"
    top_year_list = os.listdir(current_state)

    
# getting the year for all  the states
    
    for year in top_year_list:
        current_year = current_state + year + "/"
        top_file_list = os.listdir(current_year)

 # reading the json file
        
        for file in top_file_list:
            current_file = current_year + file
            data = open(current_file,'r')
            
            sixth_data = json.load(data)
            for i in sixth_data['data']['pincodes']:
                    
                    name = i['name']
                    registeredUsers = i['registeredUsers']
                    columns6['Pincode'].append(name)
                    columns6['Registered_Users'].append(registeredUsers)
                    columns6['States'].append(state)
                    columns6['Year'].append(year)
                    columns6['Quarters'].append(int(file.strip('.json')))

df_top_user= pd.DataFrame(columns6)
           
                               
```

```python
# agg_trans table creation in mysql

#columns1 = {'States': [], 'Year': [], 'Quarters':[], 'Transaction_type': [], 'Transaction_count':[], 'Transaction_amount':[] }

drop_query = "DROP TABLE IF EXISTS agg_trans"
mycursor.execute(drop_query)

create_query = '''Create table if not exists agg_trans (States varchar(100),
                                            Year int,
                                            Quarters int,
                                            Transaction_type varchar(100),
                                            Transaction_count int, 
                                            Transaction_amount double )'''

mycursor.execute (create_query) 

for i,row in df_agg_trans.iterrows():
    #here %S means string values 
    insert_query = "INSERT INTO agg_trans VALUES (%s,%s,%s,%s,%s,%s)"
    mycursor.execute(insert_query, tuple(row))

#print(mycursor.rowcount, "record(s) inserted.")
# Commit the changes after all batches are inserted
mysql.commit()                                  
                                            
```
